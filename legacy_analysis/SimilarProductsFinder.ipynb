```python
import json
cleanAsinDict = json.load(open('./out/usefulHomeKitchenCleanBOWSentimentDict.json'))
```

```python
import ast

cleanAsinSimilarProductsPriceBrandImageDict = {}
products_found = 0
lines_read = 0

with open('../amazon_yolo_data/metadata.json') as fp: 
    line = fp.readline()
    while line:
        lines_read += 1
        d = ast.literal_eval(line.strip())
        
        if d['asin'] in cleanAsinDict:
            products_found += 1
            metadataDict = {}
            similarList = []
            if 'related' in d:
                realtedDict = d['related']
                
                if 'also_bought' in realtedDict:
                    for asin in realtedDict['also_bought']:
                        if asin in cleanAsinDict:
                            similarList.append(asin)
                            
                if 'also_viewed' in realtedDict:
                    for asin in realtedDict['also_viewed']:
                        if asin in cleanAsinDict:
                            similarList.append(asin)
                            
            metadataDict["similar"] = similarList

            if 'price' in d:
                metadataDict["price"] = d['price']
            if 'brand' in d:
                metadataDict["brand"] = d['brand']
            if 'imUrl' in d:
                metadataDict["imUrl"] = d['imUrl']
            
            cleanAsinSimilarProductsPriceBrandImageDict[d['asin']] =  metadataDict
        
        if lines_read % 100000 == 0:
            print("lines read:",lines_read,"products found:", products_found)
        
        line = fp.readline()
```

```python
with open('./out/cleanAsinSimilarProductsPriceBrandImageDict.json', 'w') as outfile:
    json.dump(cleanAsinSimilarProductsPriceBrandImageDict, outfile)
```
