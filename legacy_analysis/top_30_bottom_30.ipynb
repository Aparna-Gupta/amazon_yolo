```python
'''
Dict [category] --> Dict [rank] --> List [asins]

'''

def reverse_asin_salesrank_function( data ) :
    sales_rank = "salesRank"
    count = 0
    reverse_categorywise_salesrank_asin = defaultdict( dict )
    for asin in data :
        if sales_rank not in data[asin] : continue  # missing sales rank value
        category_and_rank_dict = data[asin][ sales_rank ]
        for category, rank in category_and_rank_dict.items() :
            if rank not in reverse_categorywise_salesrank_asin [ category ] :
                reverse_categorywise_salesrank_asin [ category ][ rank ] = [] # level 1
            reverse_categorywise_salesrank_asin[ category ][ rank ].append( asin ) 
            count += 1
    return reverse_categorywise_salesrank_asin, count

reverse_categorywise_salesrank_asin, count = reverse_asin_salesrank_function( sales_data )
reverse_categorywise_salesrank_asin
```

```python
the_30_percent_limit =  round ( 0.3 * count )
the_30_percent_limit
```

```python
top_30_percent_categorywise_asins = {}
bottom_30_percent_categorywise_asins = {}
for category, single_category_dict in reverse_categorywise_salesrank_asin.items() :
    sorted_ranks = sorted( list( single_category_dict.keys() ), reverse = True )
    if sorted_ranks :
        asins = []
        for rank in sorted_ranks :
            asins.extend( single_category_dict[ rank ] ) #list of asins
            
        top_30_percent_categorywise_asins[ category ] = asins[ : round( 0.3 * len(asins)) ]
        bottom_30_percent_categorywise_asins[ category ] = asins[ -1 * round( 0.3 * len(asins)) : ]
        print("Total asins :", len(asins))
        print( "Top 30% :", len(top_30_percent_categorywise_asins[ category ] ) )
        print( "Bottom 30% :", len(bottom_30_percent_categorywise_asins[ category ] ) )
# top_30_percent_categorywise_asins
```

```python
'''
Collect top 30 and bottom 30 asin data from sales data

'''
top_30_percent_data_catogorywise = defaultdict( dict )
bottom_30_percent_data_catogorywise = defaultdict( dict )
for category, asins_list in top_30_percent_categorywise_asins.items() :
    for asin in asins_list :
        data = sales_data[ asin ]
        top_30_percent_data_catogorywise[category][ asin ] = data

for category, asins_list in bottom_30_percent_categorywise_asins.items() :
    for asin in asins_list :
        data = sales_data[ asin ]
        bottom_30_percent_data_catogorywise[category][ asin ] = data
```
