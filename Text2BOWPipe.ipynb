```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
from bs4 import BeautifulSoup

def text2BOWList(text):
    # split into words by white space
    soup = BeautifulSoup(text, "lxml")
    text = soup.get_text()
    tokens = word_tokenize(text)
    # convert to lower case
    tokens = [w.lower() for w in tokens]
    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]
    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]
    # filter out stop words
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    
    return words
```

```python
HomeKitchenCleanBOWDict = {}
for asin in HomeKitchenCleanData:
    if len(HomeKitchenCleanData[asin]['Best review']) > 0 and len(HomeKitchenCleanData[asin]['description']) > 0:
        HomeKitchenCleanBOWDict[asin] = HomeKitchenCleanData[asin]
        HomeKitchenCleanBOWDict[asin]['reviewBow'] = text2BOWList(HomeKitchenCleanData[asin]['Best review'])
        HomeKitchenCleanBOWDict[asin]['descriptionBow'] = text2BOWList(HomeKitchenCleanData[asin]['description'])    
```
