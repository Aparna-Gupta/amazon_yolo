```python
# histogram for similar product groups
import json
import pandas as pd

json_data=open('./out/HomeKitchenFinal2.json').read()
HomeKitchenFinalDict = json.loads(json_data)

json_data=open('./out/cleanAsinSimilarProductsPriceBrandImageDict.json').read()
cleanAsinSimilarProductsPriceBrandImageDict = json.loads(json_data)

# Further analysis on HomeKitchenCleanData for now
# lets get the lengths of all description and reviews in this data
goodAsins = []
for asin in cleanAsinSimilarProductsPriceBrandImageDict:
    noOfSimilarProducts = len(cleanAsinSimilarProductsPriceBrandImageDict[asin]["similar"])
    if noOfSimilarProducts == 85:
        goodAsins.append(asin)
```

```python
sample = []

asin = goodAsins[0]
sample.append(asin)

i = 0
for asin2 in cleanAsinSimilarProductsPriceBrandImageDict[asin]["similar"]:
    sample.append(asin2)

HomeKitchenFinalDict[sample[0]]
```

```python
sampleDict = {}
for asin in sample:
    sampleDict[asin] = {}
    sampleDict[asin]['reviewPoliteness'] = HomeKitchenFinalDict[asin]["reviewPoliteness"]
    sampleDict[asin]['descriptionPoliteness'] = HomeKitchenFinalDict[asin]["descriptionPoliteness"]
    
    sampleDict[asin]['reviewSentimentNeg'] = HomeKitchenFinalDict[asin]["reviewSentiment"]['neg']
    sampleDict[asin]['reviewSentimentNeu'] = HomeKitchenFinalDict[asin]["reviewSentiment"]['neu']
    sampleDict[asin]['reviewSentimentPos'] = HomeKitchenFinalDict[asin]["reviewSentiment"]['pos']
    sampleDict[asin]['reviewSentimentCom'] = HomeKitchenFinalDict[asin]["reviewSentiment"]['compound']
    
    sampleDict[asin]['descriptionSentimentNeg'] = HomeKitchenFinalDict[asin]["descriptionSentiment"]['neg']
    sampleDict[asin]['descriptionSentimentNeu'] = HomeKitchenFinalDict[asin]["descriptionSentiment"]['neu']
    sampleDict[asin]['descriptionSentimentPos'] = HomeKitchenFinalDict[asin]["descriptionSentiment"]['pos']
    sampleDict[asin]['descriptionSentimentCom'] = HomeKitchenFinalDict[asin]["descriptionSentiment"]['compound']
    
    for empath_key in HomeKitchenFinalDict[asin]["empath_description"]:
        sampleDict[asin]['empath_description_'+str(empath_key)] = HomeKitchenFinalDict[asin]["empath_description"][empath_key]
        
    for empath_key in HomeKitchenFinalDict[asin]["empath_review"]:
        sampleDict[asin]['empath_review_'+str(empath_key)] = HomeKitchenFinalDict[asin]["empath_review"][empath_key]
   
    
    sampleDict[asin]['SR'] = HomeKitchenFinalDict[asin]['salesRank']['Home &amp; Kitchen']

sampleDf = pd.DataFrame.from_dict(sampleDict, orient='index')
sampleDf.sort_values(by=['SR'])
```
