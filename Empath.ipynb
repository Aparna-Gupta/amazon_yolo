```python
from empath import Empath
```

```python
lexicon = Empath()
```

```python
import json
file_name = "out/HomeKitchenCleanBOWDict.json"
home_kitchen_BOW = json.load ( open ( file_name ) )
```

```python
home_kitchen_BOW
```

```python
review_str = home_kitchen_BOW [ 'B008UXJ21C' ][ 'reviewBow' ] 
review_str
```

```python
combined_word = " ".join( review_str )
combined_word
```

```python
lexicon.analyze( combined_word, normalize=True)
```

```python
for key in home_kitchen_BOW :
    review_str = None
    try :
        review_str = home_kitchen_BOW [ key ][ 'reviewBow' ] 
    except KeyError :
        print( "Error : " + key )
    combined_word_review = " ".join( review_str )
    output_review = lexicon.analyze( combined_word_review, normalize=True)

    home_kitchen_BOW[ key ]['empath_review'] = output_review
    
    desc_str = home_kitchen_BOW [ key ][ 'descriptionBow' ] 
    combined_word_desc = " ".join( desc_str )
    output_desc = lexicon.analyze( combined_word_desc, normalize=True)
    
    home_kitchen_BOW[ key ]['empath_description'] = output_desc
    
    
    
```

```python
with open('out/HomeKitchenCleanBOWDictEmpath.json', 'w') as outfile:
        json.dump(home_kitchen_BOW, outfile)
```
